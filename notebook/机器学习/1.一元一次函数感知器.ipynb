```python
from utils import util
from matplotlib import pyplot as plt
# numpyAPI手册: https://numpy.org/doc/stable/reference/index.html
xs, ys = util.getBeans(100)
```

```python
plt.title("Size-Toxicity Function")
plt.xlabel("size")
plt.ylabel("toxictity")
# 通过调整预测参数使函数拟合
# 设置初始系数
w = 0.5
alpha = 0.01
for i in range(100):
    y_pre = w * xs[i]
    e = ys[i] - y_pre
    w = w + e * xs[i] * alpha
y_pre = w * xs
plt.scatter(xs,ys)
plt.plot(xs,y_pre)
```
